```python
#import libraries
import pandas as pd 
import numpy as np
import string
import re
import nltk
nltk.download('all')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer



stop_words = set(stopwords.words('english'))
```

```python
#reads original dataframe
df = pd.read_csv('news.csv')
df
```

```python
df_text = pd.read_csv('news.csv', encoding='utf8')       #reads dataframe into new one and utf8 supports almost every character
df_text.columns = ['id','title', 'text', 'label']        #creates column names
df_text = df_text.drop(['id', 'title'], axis=1)          #shows only necessary columns
df_text
```

```python
#looks for urls in the 'text' column and rewords them, placing new text into a new column
df_text['cleanText'] = df_text['text'].replace(r'http\S+', '', regex=True).replace(r'www\S+', '', regex=True)
df_text
```

```python
#removes —, quotation marks, other miscellaneous symbols not in string.punctuation
df_text['cleanText'] = df_text['cleanText'].replace({'—':' ','"':'','‘':'','’':'','“':'','”':'','\n\n':' ','© ':'','':''}, regex=True)
df_text
```

```python
string.punctuation
```

```python
#removes punctuation from cleanText column
punct = "\n\r"+string.punctuation
df_text['cleanText'] = df_text['cleanText'].str.translate(str.maketrans('','',punct))
df_text
```

```python
stop_words
```

```python
#turns cleanText lowercase
df_text['cleanText'] = df_text['cleanText'].str.lower()

#removes numbers
df_text['cleanText'] = df_text['cleanText'].str.replace('\D+', '')
df_text
```

```python
#removes stopwords
df_text['cleanText'] = df_text['cleanText'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df_text
```

```python
#makes the real/fake labels binary
df_text['label'] = df_text['label'].str.replace('REAL','1')
df_text['label'] = df_text['label'].str.replace('FAKE','0')
df_text
```

```python
#makes a copy of df_text for stemming example
df_text2 = df_text.copy(deep=True).drop(['text'],axis=1)
df_text2
```

```python
#stemming example
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
ps = PorterStemmer()

def lemmatize_text(text):
    return [ps.stem(w) for w in w_tokenizer.tokenize(text)]

df_text2['stemming'] = df_text2['cleanText'].apply(lemmatize_text)
df_text2
```

```python
#dropping 'text' column
df_text = df_text.drop(['text'],axis=1)

#lemmitization and tokenizing
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

df_text['tokenized'] = df_text['cleanText'].apply(lemmatize_text)
df_text
```

```python
#tokenization
token = df_text.cleanText
token = token.apply(word_tokenize)
token
```

```python
#
DF = {}
for i in range(len(df_text['tokenized'])):
    tokens = df_text['tokenized'][i]
    for w in tokens:
        try:
            DF[w].add(i)
        except:
            DF[w] = {i}
```

```python
#counts number of times a word appears in all documents
for i in DF:
    DF[i] = len(DF[i])
DF
```
